# Outliers

```python
import pandas as pd

quant_features = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points"
]

outlier_summary = {}

for col in quant_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary[col] = {
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "outlier_count": outliers.shape[0],
        "outlier_percentage": round(outliers.shape[0] / len(df) * 100, 2)
    }

outlier_df = pd.DataFrame(outlier_summary).T
outlier_df
```

```python
import pandas as pd
import numpy as np

def iqr_outlier_ranges(df, numeric_cols, k=1.5):
    """
    For each column in numeric_cols:
      - Q1, Q3 (IQR range)
      - outlier_min, outlier_max (values outside [Q1 - k*IQR, Q3 + k*IQR])
    """
    rows = []
    for col in numeric_cols:
        s = pd.to_numeric(df[col], errors='coerce').dropna()
        if s.empty:
            rows.append({"feature": col, "Q1": np.nan, "Q3": np.nan,
                         "outlier_min": np.nan, "outlier_max": np.nan})
            continue

        q1 = s.quantile(0.25)
        q3 = s.quantile(0.75)
        iqr = q3 - q1
        lower = q1 - k * iqr
        upper = q3 + k * iqr

        outliers = s[(s < lower) | (s > upper)]
        out_min = outliers.min() if not outliers.empty else np.nan
        out_max = outliers.max() if not outliers.empty else np.nan

        rows.append({
            "feature": col,
            "Q1": q1,
            "Q3": q3,
            "outlier_min": out_min,
            "outlier_max": out_max
        })

    return pd.DataFrame(rows)[["feature","Q1","Q3","outlier_min","outlier_max"]]

quant_features = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points"
]
```

```python
summary = iqr_outlier_ranges(df, quant_features, k=1.5) 
```

```python
summary
```

```python
iqr_outlier_ranges(df, quant_features, k=3.0) 
```

# Scale of numeric

```python
import numpy as np
import pandas as pd

quant_features = [
    "Elevation",
    "Aspect",
    "Slope",
    "Horizontal_Distance_To_Hydrology",
    "Vertical_Distance_To_Hydrology",
    "Horizontal_Distance_To_Roadways",
    "Hillshade_9am",
    "Hillshade_Noon",
    "Hillshade_3pm",
    "Horizontal_Distance_To_Fire_Points"
]

cols = [c for c in quant_features if c in df.columns]
missing_cols = sorted(set(quant_features) - set(cols))
if missing_cols:
    print("Not found in df:", missing_cols)


df[cols] = df[cols].replace('?', np.nan)
for c in cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')


summary = df[cols].agg(['count','min','max','mean','std']).T
summary['missing'] = len(df) - summary['count']
summary['range'] = summary['max'] - summary['min']

# Order by widest range first
summary = summary[['min','max','range','mean','std','count','missing']] \
                 .sort_values('range', ascending=False)

summary.round(3)

```
